```python
'''PLOTS RAW EYE GAZE DATA'''
import matplotlib.pyplot as plt
import pandas as pd

# load eye tracking data
eye_data = pd.read_csv("gaze_data/P102R108387/gaze.csv")

# create scatter plot of eye gaze positions
plt.scatter(eye_data['x_position'], eye_data['y_position'], s=5, alpha = 0.5)
plt.xlabel('Horizontal position')
plt.ylabel('Vertical position')
plt.title('Raw Eye Gaze Data')


# plot the gaze path with lines and arrows
plt.plot(eye_data['x_position'], eye_data['y_position'], color='gray', alpha=0.5)
plt.arrow(eye_data['x_position'].iloc[0], eye_data['y_position'].iloc[0],
          eye_data['x_position'].iloc[1] - eye_data['x_position'].iloc[0],
          eye_data['y_position'].iloc[1] - eye_data['y_position'].iloc[0],
          length_includes_head=True, head_width=10, color='blue')
plt.arrow(eye_data['x_position'].iloc[-2], eye_data['y_position'].iloc[-2],
          eye_data['x_position'].iloc[-1] - eye_data['x_position'].iloc[-2],
          eye_data['y_position'].iloc[-1] - eye_data['y_position'].iloc[-2],
          length_includes_head=True, head_width=10, color='red')

plt.show()
```

```python
'''PLOTS TRANCRIPT WORDS'''
#Not functioning optimally
import matplotlib.pyplot as plt
from collections import Counter
import re
import string
import os

mainPath = 'main_data'  # path to the main data directory
transcriptFilePaths = []  # list to hold the paths to the transcript files

# Loop through all subdirectories in the main data directory
for subdir in os.listdir(mainPath):
    # Get the full path to the subdirectory
    subdirPath = os.path.join(mainPath, subdir)
    # Check if the subdirectory contains a transcription.txt file
    if os.path.isfile(os.path.join(subdirPath, 'transcription.txt')):
        # If it does, append the path to the transcript_files list
        transcriptFilePaths.append(os.path.join(subdirPath, 'transcription.txt'))

# Loop through the transcript_files list and read the contents of each file
allPatients = ''
counter = 0;
for filePath in transcriptFilePaths:
    counter += 1;
    with open(filePath) as f:
        allPatients += f.read()

# Remove all punctuation marks from the string
allPatients = allPatients.translate(str.maketrans('', '', string.punctuation))

# Convert all words to lowercase and split the string into a list of words
words_list = allPatients.lower().split()

# Remove certain words from patient transcipts
removeWords = ["and", "or", "no", "is", "the", "are", "in", "of"]
words_list = [word for word in words_list if word not in removeWords]

# Count the occurrence of each word in the list
word_count = Counter(words_list)

# Get the top 10 most frequent words
top_10_words = word_count.most_common(10)

# Print the word count for all words
for word, count in word_count.items():
    print(f"{word}: {count}")

# Print the top 10 most frequent words
print("Top 10 words:")
for word, count in top_10_words:
    print(f"{word}: {count}")

# Plot a histogram of the top 10 most frequent words
plt.bar([word for word, count in top_10_words],
        [count for word, count in top_10_words], width = 0.8)
plt.title("Top 10 most frequent words")
plt.xlabel("Words")
plt.ylabel("Count")

# Customize the x-axis labels
plt.xticks(rotation=90, fontsize=8)
#plt.gca().yaxis.grid(false)  # horizontal lines for the bars

plt.show()

counter
```

```python
'''PLOTS A NUMERICAL HEAT MAP'''
import pandas as pd
import seaborn as sns

# Load fixation data from CSV file
fixations = pd.read_csv("Plot Fixation Image Test Sample/p18796569/fixations.csv")

# Group fixations by their (x, y) coordinates
fixation_groups = fixations.groupby(['X_ORIGINAL', 'Y_ORIGINAL'])

# Compute the duration of each fixation group and normalize the durations to [0, 1]
max_duration = fixations["Time (in secs)"].max()
fixation_durations = fixation_groups["Time (in secs)"].sum() / max_duration

# Convert the fixation durations into a pivot table with x and y coordinates as indices
heatmap_data = fixation_durations.reset_index().pivot("Y_ORIGINAL", "X_ORIGINAL", "Time (in secs)")

# Create a heatmap using the pivot table data
sns.heatmap(heatmap_data, cmap="hot", annot=True, cbar=False)
```

```python
"""PLOTS A 3D PDF"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the fixation data
df = pd.read_csv("Plot Fixation Image Test Sample/p18796569/fixations.csv")
x = df['X_ORIGINAL']
y = df['Y_ORIGINAL']
weights = df['Time (in secs)']

# Calculate the PDF of the fixation data
xmin, xmax = x.min(), x.max()
ymin, ymax = y.min(), y.max()
nbins = 50  # adjust this value to control the resolution of the PDF
H, xedges, yedges = np.histogram2d(x, y, bins=nbins, range=[[xmin, xmax], [ymin, ymax]], weights=weights)
H /= H.sum()
X, Y = np.meshgrid(xedges[:-1], yedges[:-1])
Z = H.T

# Plot the PDF in a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('PDF')
plt.show()
```

```python
'''PLOTS 2D PDF BY COLLAPSING THE Z-AXIS OF 3D PDF'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the fixation data
df = pd.read_csv("Plot Fixation Image Test Sample/p18796569/fixations.csv")
x = df['X_ORIGINAL']
y = df['Y_ORIGINAL']
weights = df['Time (in secs)']

# Calculate the PDF of the fixation data
xmin, xmax = x.min(), x.max()
ymin, ymax = y.min(), y.max()
nbins = 50  # adjust this value to control the resolution of the PDF
H, xedges, yedges = np.histogram2d(x, y, bins=nbins, range=[[xmin, xmax], [ymin, ymax]], weights=weights)
H /= H.sum()

# Plot the PDF in a 2D plot
fig = plt.figure()
ax = fig.add_subplot(111)
ax.imshow(H.T, extent=[xmin, xmax, ymin, ymax], origin='lower', cmap='viridis', aspect='auto')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```
